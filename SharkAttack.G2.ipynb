```python
type_mapping = {
"Invalid": "Under investigation",
"Boat": "Watercraft",
" Provoked" : "Provoked",
"?": "Under investigation",
"Unconfirmed":"Under investigation",
"Unverified":"Under investigation",
"Questionable":"Under investigation"
}

df["type"] = df["type"].replace(type_mapping)
df["type"].value_counts()
```

```python
df["country"].value_counts()[df["country"].value_counts() > 30]
```

```python
# Paso 1: Contar las apariciones de cada valor Ãºnico en la columna
conteo_valores = df["country"].value_counts()

# Paso 2: Filtrar los valores que aparecen mÃ¡s veces que el nÃºmero deseado (por ejemplo, 5 veces)
valores_frecuentes = conteo_valores[conteo_valores > 30].index

# Paso 3: Filtrar las filas del DataFrame original donde la columna contiene esos valores
df = df[df["country"].isin(valores_frecuentes)]


```

```python
df["country"] = df["country"].str.replace("REUNION","REUNION ISLAND")
```
